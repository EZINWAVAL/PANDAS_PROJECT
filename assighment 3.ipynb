```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Pandas_Project.csv")

print(df)
```

```python
df.info()
```

```python
df["Amount"]=pd.to_numeric(df["Amount"].replace({r"[\$,]":""},regex=True))
```

```python
df.info()
```

```python
print(df)
```

```python
df["Date"]=pd.to_datetime(df["Date"],errors="coerce")
```

```python
df["month"]=df["Date"].dt.month_name()
```

```python
print(df)
```

```python
df["Amount"]=pd.to_numeric(df["Amount"].replace({r"[\$,]":""},regex=True))
```

```python
highest_sales_person=df.groupby("Sales Person")["Amount"].sum()
```

```python
highest_sales_person
```

```python
highest_sales_person.sort_values(ascending=False)
```

```python
product_per_boxes=df.groupby("Product")["Boxes Shipped"].count()
```

```python
product_per_boxes
```

```python
product_per_boxes.sort_values(ascending=False)
```

```python
Total_sales_per_month=df.groupby("month")["Amount"].sum()
```

```python
Total_sales_per_month
```

```python
Total_sales_per_month.sort_values(ascending=False)
```

```python
Total_sales_per_month.index=pd.to_datetime(Total_sales_per_month.index,format="%B")
```

```python
Total_sales_per_month=Total_sales_per_month.sort_index()
```

```python
Total_sales_per_month
```

```python
df["revenue_per_box"]=df["Amount"]/ df["Boxes Shipped"]
```

```python
df["revenue_per_box"]=df["Amount"]/ df["Boxes Shipped"].mean()
```

```python
df["revenue_per_box"]
```

```python
revenue_per_country=df.groupby("Country")["Amount"].sum()
```

```python
revenue_per_country
```

```python
revenue_per_country.sort_values(ascending=False)
```

```python
average_revenue_product=df.groupby("Product")["Amount"].mean()
```

```python
average_revenue_product
```
